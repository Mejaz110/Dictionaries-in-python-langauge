```python
# Parent Class
class BankAccount:
    def __init__(self, name, account_no, balance):
        self.name = name
        self.account_no = account_no
        self.balance = balance


# Child Class 1: Saving Account
class SavingAccount(BankAccount):
    def show_balance(self):
        print(f"Account Holder: {self.name}")
        print(f"Account No: {self.account_no}")
        print(f"Saving Account Balance: {self.balance}")

    def add_interest(self, interest_rate):
        interest = self.balance * (interest_rate / 100)
        self.balance += interest
        print(f"Interest Added: {interest}")
        print(f"New Balance: {self.balance}")


# Child Class 2: Current Account
class CurrentAccount(BankAccount):
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawn: {amount}")
            print(f"Remaining Balance: {self.balance}")
        else:
            print("Insufficient balance!")





```

```python
# Saving Account object
s1 = SavingAccount("Ali", 101, 2000)
s1.show_balance()
s1.add_interest(5)  # 5% interest

```

```python
# Current Account object
c1 = CurrentAccount("Ahmed", 202, 5000)
c1.withdraw(1500)

```
